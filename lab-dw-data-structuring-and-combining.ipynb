This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.drop(columns=["unnamed:_0"])
```

```python
pivot_revenue = (df.pivot_table(values="total_claim_amount",index="sales_channel",aggfunc="sum").round(2))

pivot_revenue
```

Sales generated through agents represent the largest share of total revenue, suggesting that personalized, agent-led interactions are currently the most effective sales channel. Branch offices also perform strongly, while digital channels such as web contribute less to total revenue, indicating potential room for growth in online sales strategies.

```python
pivot_clv = (df.pivot_table(values="customer_lifetime_value",index="gender",columns="education",aggfunc="mean").round(2))

pivot_clv
```

Average customer lifetime value varies more strongly by education level than by gender. Customers with lower formal education (“High School or Below”) show the highest CLV on average, while customers with doctoral education present the lowest values. Gender differences are minimal across all education levels, suggesting that education is a more relevant factor than gender in explaining long-term customer value.

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
df["number_of_open_complaints"].value_counts()
```

```python
complaints_long = (df.groupby(["policy_type", "month"])["number_of_open_complaints"].sum().reset_index())

complaints_long
```

```python
df["month"].unique()
```

The analysis shows that complaint volume differs significantly by policy type. Personal Auto consistently registers the highest number of complaints in both observed months, making it the most complaint-prone policy category. Corporate Auto presents a moderate level of complaints, while Special Auto shows very low complaint volumes across the period.

From a temporal perspective, most policy types display a higher number of complaints in month 1 compared to month 2, suggesting a potential short-term seasonal or operational pattern. An exception is Special Auto, which shows a slight increase in complaints in month 2, although the overall number remains low.

It is important to note that the dataset only includes data for two months. Therefore, conclusions about seasonality are limited to this time range and should be interpreted with caution.

Overall, these results suggest that customer service efforts could be prioritized toward Personal Auto policies, particularly during periods with higher complaint volumes.
